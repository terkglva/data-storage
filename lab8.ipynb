```python
# Part A: Conceptual Questions
print("=" * 60)
print("PART A: CONCEPTUAL QUESTIONS")
print("=" * 60)

print("""
1. Main differences between relational (SQL) and non-relational (NoSQL) databases:
   - Relational databases use structured tables with fixed schemas, support relationships 
     through keys, and use SQL for queries. NoSQL databases have flexible schemas, can 
     store unstructured data, and scale horizontally.
   - SQL example: Banking system where data integrity and complex transactions are critical
   - NoSQL example: Logging system with high-volume data and evolving structure

2. For an online bookstore, I recommend a relational database because:
   - Data has clear structure with multiple relationships (books-authors, orders-customers)
   - Data integrity is crucial for order processing
   - Complex queries are needed for sales analytics
   - ACID transactions are required for order management

3. ACID properties:
   - Atomicity: Transaction completes fully or not at all
   - Consistency: Data remains valid after transaction
   - Isolation: Concurrent transactions don't interfere
   - Durability: Committed changes persist
   Important for orders to prevent lost orders or double charges during failures

4. Primary key - unique identifier for each record (e.g., ISBN for books)
   Foreign key - references primary key in another table (e.g., customer_id in Orders table)

5. Data redundancy leads to:
   - Wasted storage space
   - Risk of data inconsistency
   - Difficult updates
   Normalization helps by organizing data into logical tables and eliminating duplication
""")
```

```python
# Part B: Database Schema
print("\n" + "=" * 60)
print("PART B: DATABASE SCHEMA")
print("=" * 60)

print("""
Relational Schema:

1. Authors
   - author_id INTEGER PRIMARY KEY
   - name VARCHAR(100) NOT NULL
   - birth_year INTEGER

2. Books
   - isbn VARCHAR(13) PRIMARY KEY
   - title VARCHAR(200) NOT NULL
   - publisher VARCHAR(100)
   - publication_year INTEGER
   - price DECIMAL(10,2) NOT NULL
   - stock_quantity INTEGER NOT NULL DEFAULT 0

3. BookAuthors (Junction Table)
   - book_isbn VARCHAR(13) FOREIGN KEY REFERENCES Books(isbn)
   - author_id INTEGER FOREIGN KEY REFERENCES Authors(author_id)
   - PRIMARY KEY (book_isbn, author_id)

4. Customers
   - customer_id INTEGER PRIMARY KEY
   - name VARCHAR(100) NOT NULL
   - email VARCHAR(100) UNIQUE NOT NULL
   - address TEXT
   - registration_date DATE NOT NULL

5. Orders
   - order_id INTEGER PRIMARY KEY
   - customer_id INTEGER FOREIGN KEY REFERENCES Customers(customer_id)
   - order_date DATE NOT NULL
   - total_amount DECIMAL(10,2)
   - status VARCHAR(20) CHECK (status IN ('pending', 'shipped', 'delivered', 'cancelled'))

6. OrderItems
   - order_item_id INTEGER PRIMARY KEY
   - order_id INTEGER FOREIGN KEY REFERENCES Orders(order_id)
   - book_isbn VARCHAR(13) FOREIGN KEY REFERENCES Books(isbn)
   - quantity INTEGER NOT NULL
   - price_at_time DECIMAL(10,2) NOT NULL
""")
```

```python
# Install: pip install matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch

fig, ax = plt.subplots(1, 1, figsize=(20, 12))
ax.set_xlim(0, 20)
ax.set_ylim(0, 14)
ax.axis('off')

# Function to draw table boxes
def draw_table(x, y, title, columns, pk_cols=None, fk_cols=None):
    if pk_cols is None:
        pk_cols = []
    if fk_cols is None:
        fk_cols = []
    
    # Calculate box height
    box_height = 1.5 + len(columns) * 0.4
    box_width = 3.5
    
    # Draw box
    rect = FancyBboxPatch((x, y), box_width, box_height, 
                          boxstyle="round,pad=0.1", 
                          facecolor='lightblue', edgecolor='black', 
                          linewidth=2)
    ax.add_patch(rect)
    
    # Draw title
    ax.text(x + box_width/2, y + box_height - 0.25, title, 
            fontsize=12, fontweight='bold', ha='center')
    
    # Draw columns
    for i, col in enumerate(columns):
        y_pos = y + box_height - 0.7 - i * 0.4
        
        # Highlight PK and FK
        if col in pk_cols and col in fk_cols:
            col_text = f"🔑🔗 {col}"
            color = 'purple'
        elif col in pk_cols:
            col_text = f"🔑 {col}"
            color = 'red'
        elif col in fk_cols:
            col_text = f"🔗 {col}"
            color = 'blue'
        else:
            col_text = f"   {col}"
            color = 'black'
        
        ax.text(x + 0.2, y_pos, col_text, fontsize=9, color=color, 
                fontfamily='monospace')

# Draw all tables
draw_table(0, 8, "AUTHORS", 
           ['author_id', 'name', 'birth_year'], 
           pk_cols=['author_id'])

draw_table(5, 8, "BOOKAUTHORS", 
           ['book_isbn', 'author_id'], 
           pk_cols=['book_isbn', 'author_id'], 
           fk_cols=['book_isbn', 'author_id'])

draw_table(10, 8, "BOOKS", 
           ['isbn', 'title', 'publisher', 'publication_year', 'price', 'stock_quantity'], 
           pk_cols=['isbn'])

draw_table(0, 2, "CUSTOMERS", 
           ['customer_id', 'name', 'email', 'address', 'registration_date'], 
           pk_cols=['customer_id'])

draw_table(5, 2, "ORDERS", 
           ['order_id', 'customer_id', 'order_date', 'total_amount', 'status'], 
           pk_cols=['order_id'], 
           fk_cols=['customer_id'])

draw_table(10, 2, "ORDERITEMS", 
           ['order_item_id', 'order_id', 'book_isbn', 'quantity', 'price_at_time'], 
           pk_cols=['order_item_id'], 
           fk_cols=['order_id', 'book_isbn'])

# Draw relationships
def draw_arrow(x1, y1, x2, y2, label='', style='->'):
    arrow = FancyArrowPatch((x1, y1), (x2, y2), 
                            arrowstyle=style, 
                            color='gray', 
                            linewidth=1.5,
                            connectionstyle="arc3,rad=0.1")
    ax.add_patch(arrow)
    if label:
        mx = (x1 + x2) / 2
        my = (y1 + y2) / 2
        ax.text(mx, my, label, fontsize=8, color='gray', 
                bbox=dict(boxstyle="round,pad=0.2", facecolor='white', edgecolor='none'))

# Draw relationship lines
draw_arrow(3.5, 9.5, 5, 9.5, 'N')
draw_arrow(5, 9.5, 3.5, 9.5, '1')
draw_arrow(8.5, 9.5, 10, 9.5, 'N')
draw_arrow(10, 9.5, 8.5, 9.5, '1')
draw_arrow(3.5, 5.5, 5, 5.5, 'N')
draw_arrow(5, 5.5, 3.5, 5.5, '1')
draw_arrow(8.5, 5.5, 10, 5.5, 'N')
draw_arrow(10, 5.5, 8.5, 5.5, '1')

# Add legend
legend_x = 15
legend_y = 1
ax.text(legend_x, legend_y + 2.5, "LEGEND:", fontweight='bold', fontsize=11)
ax.text(legend_x, legend_y + 2.0, "🔑 Primary Key", fontsize=9, color='red')
ax.text(legend_x, legend_y + 1.6, "🔗 Foreign Key", fontsize=9, color='blue')
ax.text(legend_x, legend_y + 1.2, "🔑🔗 Composite PK/FK", fontsize=9, color='purple')
ax.text(legend_x, legend_y + 0.8, "1 : One", fontsize=9)
ax.text(legend_x, legend_y + 0.4, "N : Many", fontsize=9)
ax.text(legend_x, legend_y + 0.0, "→ Relationship", fontsize=9)

plt.title("ENTITY-RELATIONSHIP DIAGRAM\nOnline Bookstore Database", 
          fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('er_diagram_matplotlib.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ ER diagram saved as 'er_diagram_matplotlib.png'")
```

```python
# Part C: SQL DDL - Creating Tables
print("\n" + "=" * 60)
print("PART C: CREATING TABLES (DDL)")
print("=" * 60)

import sqlite3
import pandas as pd
from datetime import datetime, timedelta

# Create database connection
conn = sqlite3.connect('bookstore.db')
cursor = conn.cursor()

# Drop existing tables
cursor.execute("DROP TABLE IF EXISTS OrderItems")
cursor.execute("DROP TABLE IF EXISTS Orders")
cursor.execute("DROP TABLE IF EXISTS Customers")
cursor.execute("DROP TABLE IF EXISTS BookAuthors")
cursor.execute("DROP TABLE IF EXISTS Books")
cursor.execute("DROP TABLE IF EXISTS Authors")

# Create Authors table
cursor.execute('''
CREATE TABLE Authors (
    author_id INTEGER PRIMARY KEY AUTOINCREMENT,
    name VARCHAR(100) NOT NULL,
    birth_year INTEGER
)
''')

# Create Books table
cursor.execute('''
CREATE TABLE Books (
    isbn VARCHAR(13) PRIMARY KEY,
    title VARCHAR(200) NOT NULL,
    publisher VARCHAR(100),
    publication_year INTEGER,
    price DECIMAL(10,2) NOT NULL,
    stock_quantity INTEGER NOT NULL DEFAULT 0
)
''')

# Create BookAuthors junction table
cursor.execute('''
CREATE TABLE BookAuthors (
    book_isbn VARCHAR(13),
    author_id INTEGER,
    PRIMARY KEY (book_isbn, author_id),
    FOREIGN KEY (book_isbn) REFERENCES Books(isbn) ON DELETE CASCADE ON UPDATE CASCADE,
    FOREIGN KEY (author_id) REFERENCES Authors(author_id) ON DELETE CASCADE ON UPDATE CASCADE
)
''')

# Create Customers table
cursor.execute('''
CREATE TABLE Customers (
    customer_id INTEGER PRIMARY KEY AUTOINCREMENT,
    name VARCHAR(100) NOT NULL,
    email VARCHAR(100) UNIQUE NOT NULL,
    address TEXT,
    registration_date DATE NOT NULL
)
''')

# Create Orders table
cursor.execute('''
CREATE TABLE Orders (
    order_id INTEGER PRIMARY KEY AUTOINCREMENT,
    customer_id INTEGER NOT NULL,
    order_date DATE NOT NULL,
    total_amount DECIMAL(10,2),
    status VARCHAR(20) CHECK (status IN ('pending', 'shipped', 'delivered', 'cancelled')),
    FOREIGN KEY (customer_id) REFERENCES Customers(customer_id) ON DELETE RESTRICT ON UPDATE CASCADE
)
''')

# Create OrderItems table
cursor.execute('''
CREATE TABLE OrderItems (
    order_item_id INTEGER PRIMARY KEY AUTOINCREMENT,
    order_id INTEGER NOT NULL,
    book_isbn VARCHAR(13) NOT NULL,
    quantity INTEGER NOT NULL,
    price_at_time DECIMAL(10,2) NOT NULL,
    FOREIGN KEY (order_id) REFERENCES Orders(order_id) ON DELETE CASCADE ON UPDATE CASCADE,
    FOREIGN KEY (book_isbn) REFERENCES Books(isbn) ON DELETE RESTRICT ON UPDATE CASCADE
)
''')

print("✓ Tables created successfully")
conn.commit()
```

```python
# Part E: SQL Queries
print("\n" + "=" * 60)
print("PART E: SQL QUERIES")
print("=" * 60)

# Query 1: List all books with their title, author(s), and price
print("\n" + "-" * 40)
print("1. All books with authors and price:")
print("-" * 40)

query1 = '''
SELECT b.title, GROUP_CONCAT(a.name, ', ') as authors, b.price
FROM Books b
JOIN BookAuthors ba ON b.isbn = ba.book_isbn
JOIN Authors a ON ba.author_id = a.author_id
GROUP BY b.isbn, b.title, b.price
ORDER BY b.title
'''
df1 = pd.read_sql_query(query1, conn)
print(df1.to_string(index=False))

# Query 2: Find all customers who registered in 2023
print("\n" + "-" * 40)
print("2. Customers registered in 2023:")
print("-" * 40)

query2 = '''
SELECT name, email, registration_date
FROM Customers
WHERE strftime('%Y', registration_date) = '2023'
'''
df2 = pd.read_sql_query(query2, conn)
print(df2.to_string(index=False))

# Query 3: Show pending orders from the last 30 days
print("\n" + "-" * 40)
print("3. Pending orders from the last 30 days (from 2024-03-20):")
print("-" * 40)

query3 = '''
SELECT o.order_id, c.name, o.order_date, o.total_amount, o.status
FROM Orders o
JOIN Customers c ON o.customer_id = c.customer_id
WHERE o.status = 'pending' 
  AND o.order_date >= date('2024-03-20', '-30 days')
ORDER BY o.order_date DESC
'''
df3 = pd.read_sql_query(query3, conn)
print(df3.to_string(index=False) if not df3.empty else "No pending orders in the last 30 days")

# Query 4: Count how many books each author has written
print("\n" + "-" * 40)
print("4. Number of books written by each author:")
print("-" * 40)

query4 = '''
SELECT a.name, COUNT(ba.book_isbn) as books_count
FROM Authors a
LEFT JOIN BookAuthors ba ON a.author_id = ba.author_id
GROUP BY a.author_id, a.name
ORDER BY books_count DESC
'''
df4 = pd.read_sql_query(query4, conn)
print(df4.to_string(index=False))

# Query 5: Total number of orders placed by each customer
print("\n" + "-" * 40)
print("5. Total orders by customer (sorted descending):")
print("-" * 40)

query5 = '''
SELECT c.name, COUNT(o.order_id) as order_count
FROM Customers c
LEFT JOIN Orders o ON c.customer_id = o.customer_id
GROUP BY c.customer_id, c.name
ORDER BY order_count DESC
'''
df5 = pd.read_sql_query(query5, conn)
print(df5.to_string(index=False))

# Query 6: Calculate total revenue from all orders
print("\n" + "-" * 40)
print("6. Total revenue from all orders:")
print("-" * 40)

query6 = '''
SELECT SUM(quantity * price_at_time) as total_revenue
FROM OrderItems
'''
df6 = pd.read_sql_query(query6, conn)
print(f"Total Revenue: ${df6['total_revenue'].iloc[0]:.2f}")

# Query 7: Find top 3 best-selling books by quantity sold
print("\n" + "-" * 40)
print("7. Top 3 best-selling books:")
print("-" * 40)

query7 = '''
SELECT b.title, SUM(oi.quantity) as total_sold
FROM Books b
JOIN OrderItems oi ON b.isbn = oi.book_isbn
GROUP BY b.isbn, b.title
ORDER BY total_sold DESC
LIMIT 3
'''
df7 = pd.read_sql_query(query7, conn)
print(df7.to_string(index=False))

# Query 8: List customers who spent more than $100
print("\n" + "-" * 40)
print("8. Customers who spent more than $100:")
print("-" * 40)

query8 = '''
SELECT c.name, SUM(oi.quantity * oi.price_at_time) as total_spent
FROM Customers c
JOIN Orders o ON c.customer_id = o.customer_id
JOIN OrderItems oi ON o.order_id = oi.order_id
GROUP BY c.customer_id, c.name
HAVING total_spent > 100
ORDER BY total_spent DESC
'''
df8 = pd.read_sql_query(query8, conn)
print(df8.to_string(index=False) if not df8.empty else "No customers spent more than $100")

# Query 9: Monthly orders and sales for 2024
print("\n" + "-" * 40)
print("9. Monthly orders and sales for 2024:")
print("-" * 40)

query9 = '''
SELECT 
    strftime('%Y-%m', order_date) as month,
    COUNT(order_id) as order_count,
    SUM(total_amount) as total_sales
FROM Orders
WHERE strftime('%Y', order_date) = '2024'
GROUP BY strftime('%Y-%m', order_date)
ORDER BY month
'''
df9 = pd.read_sql_query(query9, conn)
print(df9.to_string(index=False))
```

```python
# Additional Queries for Verification
print("\n" + "-" * 40)
print("ADDITIONAL ANALYSIS:")
print("-" * 40)

# Books with multiple authors
query_multiple = '''
SELECT b.title, COUNT(ba.author_id) as author_count
FROM Books b
JOIN BookAuthors ba ON b.isbn = ba.book_isbn
GROUP BY b.isbn, b.title
HAVING author_count > 1
'''
df_multiple = pd.read_sql_query(query_multiple, conn)
print("\nBooks with multiple authors:")
print(df_multiple.to_string(index=False))

# Order status statistics
query_status = '''
SELECT status, COUNT(*) as count, SUM(total_amount) as total
FROM Orders
GROUP BY status
'''
df_status = pd.read_sql_query(query_status, conn)
print("\nOrder status statistics:")
print(df_status.to_string(index=False))

# Inventory status
query_inventory = '''
SELECT title, stock_quantity, price
FROM Books
ORDER BY stock_quantity DESC
'''
df_inventory = pd.read_sql_query(query_inventory, conn)
print("\nCurrent inventory status:")
print(df_inventory.to_string(index=False))
```
